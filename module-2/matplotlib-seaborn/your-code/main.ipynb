```python
# tamaño
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))

# Título
fig.suptitle("Comparación de gráficos", fontsize=10, fontweight='bold')

axs[0].plot(x, y)
axs[0].set_title('lineal', fontsize=10, fontweight='bold')
axs[0].set_xlabel('X', fontsize=10)
axs[0].set_ylabel('Y', fontsize=10)

axs[1].plot(x, z)
axs[1].set_title('exponencial', fontsize=10, fontweight='bold')
axs[1].set_xlabel('X', fontsize=10)
axs[1].set_ylabel('Z', fontsize=10)

# Ajustar el diseño
plt.subplots_adjust(wspace=0.5)

# Mostrar los gráficos
plt.show()
```
